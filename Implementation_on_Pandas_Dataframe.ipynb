**Pipe Function**

1. pipe function # function defination where it will add or sub with entire table

```python
data = {'One' : pd.Series([1,2,3,4,5]),
       'Two' : pd.Series([10,20,30,40,50]),
       'Three' : pd.Series([11,21,34,45,58]),
       'Four' : pd.Series([100,200,300,400,500])}

def add_(i,j):
  return i + j

def sub_(i,j):
  return i - j
df = pd.DataFrame(data)
df
```

```python
df.pipe(add_,10)
```

```python
df.pipe(sub_,10)
```

**Groupby Function**

```python
cricket = {'Team' : ['india','india','australia','australia','sa','sa','sa','sa','nz','nz','nz','india'],
           'rank' : [2,3,1,2,3,4,1,1,2,4,1,2],
           'year' : [2014,2015,2012,2015,2016,2017,2014,2017,2018,2012,2015,2015],
           'points': [812,678,789,876,789,456,765,976,786,467,567,898]}
df = pd.DataFrame(cricket)
df
```

```python
df.groupby('Team').groups
```

serach by specific group

```python
df.groupby(['Team','year']).get_group(('australia',2012))
```

if data is not there means it will display error

adding some statistical computation on top of groupby

```python
df.groupby('Team')['points'].sum()
```

```python
df.groupby('Team').sum()['points'].sort_values(ascending = False)
```

To check multiple stats for points team wise

```python
groups = df.groupby('Team')
groups['points'].agg([np.sum,np.mean,np.std])
```

filter function along with groupby

```python
df.groupby('Team').filter(lambda x: len(x) == 4)
```
